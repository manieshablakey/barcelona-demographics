### Barcelona data - 2017

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Population by gender
pop = pd.read_csv("exports/1-pop-by-gender.csv")
pop_cat = np.arange(len(pop["Gender"]))
xpop = pop["Gender"]
ypop = pop["Total Population"]
 
plt.bar(pop_cat, ypop, alpha=0.8)
plt.xticks(pop_cat, xpop)
plt.ylabel('Total Population')
plt.title('Total population in 2017')
plt.show()
```

```python
#births by gender
births = pd.read_csv("exports/2-births-by-gender.csv")
birth_cat = np.arange(len(births["Gender"]))
xbirth = births["Gender"]
ybirth = births["Total Births"]
 
plt.bar(birth_cat, ybirth, alpha=0.8)
plt.xticks(birth_cat, xbirth)
plt.ylabel('Total Births')
plt.title('Births by gender')
plt.show()
```

```python
#deaths by age
deaths = pd.read_csv("exports/3-deaths-by-age.csv")
# deaths

deaths_cat = np.arange(len(deaths["Age"]))
xdeaths = deaths["Age"]
ydeaths = deaths["Total Deaths"]
 
plt.bar(deaths_cat, ydeaths, alpha=0.8, color="g")
plt.xticks(deaths_cat, xdeaths)
plt.ylabel('Total Deaths')
plt.title('Deaths by age')
plt.show()
```

```python
# immigrants by nationality
imm_nation = pd.read_csv("exports/5-immigrants-by-nationality.csv")

nation = imm_nation["Nationality"]
nation = nation.tolist()

immigrants = imm_nation["Total immigrants"]
immigrants = immigrants.tolist()

imm_nation = np.arange(len(nation))
ximm_nation = nation
yimm_nation = immigrants
 
plt.bar(imm_nation, yimm_nation, alpha=0.8)
plt.xticks(imm_nation, ximm_nation)
plt.ylabel('Total Immigrants')
plt.title('Countries where most immigrants come from (Top 10)')
plt.show()
```
